```python
import numpy as np

def quad():
    print("Quadratic equation: ax^2 + bx + c = 0")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    roots = np.roots([a, b, c])
    print("Roots are:", roots)
def cubic():
    print("Cubic equations: ax^3 + bx^2 + cx + d = 0")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))
    roots = np.roots([a, b, c, d])
    print("Roots are:", roots)
def quart():
    print("Quartic equation: ax^4 + bx^3 + cx^2 + dx + e = 0")
    a = float(("Enter a: "))
    b = float(("Enter b: "))
    c = float(("Enter c: "))
    d = float(("Enter d: "))
    e = float(("Enter e: "))
    roots = np.roots([a, b, c, d, e])
    print("Roots are:", roots)
while True:
    print("\nChoose the equation to solve:")
    print("1. Quadratic")
    print("2. Cubic")
    print("3. Quartic")
    print("4. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        quad()
    elif choice == "2":
        cubic()
    elif choice == "3":
        quart()
    elif choice == "4":
        print("Existing program...")
        break
    else:
        print("Invalid choice.")
```
